```python
import pandas as pd
from pytrends.request import TrendReq
import matplotlib.pyplot as plt
```

```python
trends = TrendReq()
```

```python
keyword = input("Enter the keyword to analyse : ")
print(f"The keyword is : {keyword}")
```

```python
trends.build_payload(kw_list=[keyword])
data = trends.interest_by_region()
data.rename(columns={keyword: "keyword"}, inplace=True)
data = data.sort_values(by="keyword", ascending=False)
region_data = data.copy()
data = data.head(20)
```

```python
data.head()
```

```python
data.reset_index().plot(x="geoName", y="keyword", figsize=(20,7), kind="bar",)
plt.title(keyword)
# plt.style.use('fivethirtyeight')
plt.show()
```

```python
data = TrendReq(hl='en-US', tz=360)
data.build_payload(kw_list=[keyword])
data = data.interest_over_time()
data.rename(columns={keyword: "keyword"}, inplace=True)
fig, ax = plt.subplots(figsize=(20, 7))
time_data = data.copy()
data['keyword'].plot()
plt.style.use('fivethirtyeight')
plt.title(f'Total Google Searches for {keyword}', fontweight='bold')
plt.xlabel('Year')
plt.ylabel('Total Count')
plt.show()
```

```python
rd = region_data.reset_index()
td = time_data.reset_index()
```

```python
rd_graph = {
    "entity_name": "Region-wise interest",
    "labels": rd["geoName"].tolist(),
    "series": rd["keyword"].tolist(),
    "property": "Google Search Analytics Region-wise",
}

td_graph = {
    "entity_name": "Trend over years",
    "labels": td["date"].dt.strftime("%Y-%m-%d").tolist(),
    "series": td["keyword"].tolist(),
    "property": "Google Search Analytics previous trends",
}
```
